```python
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
import os
import random
```

```python
total=os.listdir('images/')
len(total)
```

```python
plt.figure(figsize=(10,10))
for i in range(10):
    plt.subplot(4,4,i+1)
    idx=random.randint(0,len(total))
    plt.imshow(cv2.imread(os.path.join('images/',total[idx])))
```

```python
import pandas as pd
import shutil
```

```python
train=pd.read_csv('training.csv')
test=pd.read_csv('test.csv')
```

```python
for im in train.image_name.values:
    
    oldpath=os.path.join('images',im)
    newpath=os.path.join('train',im)
    shutil.copy2(oldpath,newpath)
```

```python
for im in test.image_name.values:
    
    oldpath=os.path.join('images',im)
    newpath=os.path.join('test',im)
    shutil.copy2(oldpath,newpath)
```
